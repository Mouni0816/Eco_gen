# Code generation

```python
from statistics import stdev
import numpy as np
import statistics
```

```python
date_list= [ '1870-1879','1880-1889','1890-1900','1900-1909','1910-1919','1920-1929','1930-1939','1940-1949','1950-1959', '1960-1969','1970-1979','1980-1989', '1990-1999','2000-2009','2010-2019']
date_list
```

```python
list_colone = ['Natural Gas Use (kBtu)',"Gross Floor Area - Buildings (sq m)",'Electricity Use (kBtu)']
```

```python
def generate_df(nb_time, list_colone, date_list):
    df_generate = pd.DataFrame(columns =["Year Built agg",'Natural Gas Use (kBtu)',"Gross Floor Area - Buildings (sq m)",'Electricity Use (kBtu)'])
    for i in range(0, nb_time):
        for date in date_list:
            df_choisi = df_office[df_office["Year Built agg"] == date]
            ligne = {}
            for colone in list_colone:
                threshold = min(df_choisi[colone])
                quartile_25 = np.percentile(df_choisi[colone], 25)
                quartile_75 = np.percentile(df_choisi[colone], 75)
                mean = statistics.mean([quartile_25, quartile_75]) 
                mu = mean
                sigma = mean - quartile_25
                generate = np.random.normal(mu, sigma, 1)
                if(generate < threshold):
                    break
                else:
                    ligne[colone] = generate[0]
            ligne["Year Built agg"] = date
            df_generate = df_generate.append(ligne, ignore_index=True)
    return df_generate.dropna()
```
